```python
import pandas as pd
```

```python
s1 = pd.Series(x, name = "Time")
```

```python
s2 = pd.Series(y, name = "speed")
```

```python
s1
```

```python
s2
```

```python
df = pd.concat([s1,s2],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures()
```

```python
x1 = pf.fit_transform(df[['Time']])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x1,df.speed)
```

```python
coef = model.coef_
coef
```

```python
intercept = model.intercept_
intercept
```

```python
y_predicted = intercept + line*coef[1] + coef[2]*line**2
```

```python
y_predicted 
```

```python
plt.scatter(x,y)
plt.plot(line,y_predicted , color='r')
plt.show()
```

```python
model.predict(x1)
```

```python
model.score(x1,df['speed'])
```
